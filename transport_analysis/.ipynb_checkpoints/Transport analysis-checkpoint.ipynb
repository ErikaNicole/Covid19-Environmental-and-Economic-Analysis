# Transport Analysis

First start with the changes in transport usage compared with the stringency level in Great Britian

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.collections as collections
plt.rcParams['figure.figsize'] = [16, 10]
```

## Use of Transport Modes

### Import transport data

```python
transport_data = pd.read_excel('data/transport_compared_february.xlsx', sheet_name='Percentage')
```

```python
x = transport_data['Date']

y_cars = transport_data['Cars']
y_national_rail = transport_data['National Rail']
y_tube = transport_data['Transport for London Tube']
y_lbus = transport_data['Transport for London Bus']
y_bus = transport_data['Bus (excl. London)']
y_cycle = transport_data['Cycling']
```

### Plot Transport Data

```python
plt.plot(x, y_cars, label='Cars')
plt.plot(x, y_national_rail, label='National Rail')
plt.plot(x, y_tube, label='Transport for London Tube')
plt.plot(x, y_lbus, label='Transport for London Bus')
plt.plot(x, y_bus, label='Bus (excl. London)')
#plt.plot(transport_data['Date'], transport_data['Cycling'], label='Cycling')

plt.axvline(x = np.datetime64('2020-03-23'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-03-24'), 100, '23rd March 2020: \n Lockdown Starts')

plt.axvline(x = np.datetime64('2020-05-13'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-05-14'), 2, '1st Lockdown Easing')

plt.axvline(x = np.datetime64('2020-06-01'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-06-02'), 6, '2nd Lockdown Easing')

plt.axvline(x = np.datetime64('2020-06-15'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-06-16'), 10, '3rd Lockdown Easing')

plt.axvline(x = np.datetime64('2020-07-04'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-07-05'), 14, '4th Lockdown Easing')

plt.xlabel('Date')
plt.ylabel('Use of Transport Modes: \n Percentage of a Equilvalent Day or Week (%)')
plt.title("Use of Transport Modes in the UK since 1st March 2020")
plt.legend()
```

# Stringency Data

### Import Stringency Data

```python
stringency_data = pd.read_excel('data/index_stringency_clean.xlsx')
```

### Plot Stringency Data

```python
plt.plot(stringency_data['Date'], stringency_data['Stringency'])
plt.axvline(x = np.datetime64('2020-03-23'), color='black', linestyle = '--')

plt.text(np.datetime64('2020-03-24'), 1, '23rd March 2020: Lockdown Starts')

plt.axvline(x = np.datetime64('2020-05-13'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-05-14'), 5, '1st Lockdown Easing')

plt.axvline(x = np.datetime64('2020-06-01'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-06-02'), 10, '2nd Lockdown Easing')

plt.axvline(x = np.datetime64('2020-06-15'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-06-16'), 15, '3rd Lockdown Easing')

plt.axvline(x = np.datetime64('2020-07-04'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-07-05'), 20, '4th Lockdown Easing')

plt.xlabel('Date')
plt.ylabel('Stringency Index in the UK (0-100)')
plt.title("Stringency Index in the UK")
```

## Use of Transport Modes vs Stringency Index

```python
fig, ax1 = plt.subplots()

ax1.plot(x, y_cars, label='Cars')
ax1.plot(x, y_national_rail, label='National Rail')
ax1.plot(x, y_tube, label='Transport for London Tube')
ax1.plot(x, y_lbus, label='Transport for London Bus')
ax1.plot(x, y_bus, label='Bus (excl. London)')
#plt.plot(x, transport_data['Cycling'], label='Cycling')
ax1.set_xlabel('Date')
ax1.set_ylabel('Use of Transport Modes: \n Percentage of a Equilvalent Day or Week (%)', color='purple')
ax1.tick_params(axis='y', labelcolor='purple')
ax1.legend()

ax2 = ax1.twinx()

ax2.plot(stringency_data['Date'], stringency_data['Stringency'], color='black')
ax2.set_ylabel('Stringency Index in the UK (0-100)')

fig.tight_layout()
plt.title("Use of Transport Modes in the UK since 1st March 2020 vs Stringency Index in the UK")
plt.axvline(x = np.datetime64('2020-03-23'), color='black', linestyle = '--')

plt.axvline(x = np.datetime64('2020-05-13'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-05-14'), -2, '1st Lockdown Easing')

plt.axvline(x = np.datetime64('2020-06-01'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-06-02'), 0, '2nd Lockdown Easing')

plt.axvline(x = np.datetime64('2020-06-15'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-06-16'), 2, '3rd Lockdown Easing')

plt.axvline(x = np.datetime64('2020-07-04'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-07-05'), 4, '4th Lockdown Easing')

plt.text(np.datetime64('2020-02-19'), -2, '23rd March 2020: \n Lockdown Starts')
plt.show()
```

#### Cycling

```python
fig, ax1 = plt.subplots()

ax1.plot(transport_data['Date'], transport_data['Cycling'], label='Cycling', color='purple')
ax1.set_xlabel('Date')
ax1.set_ylabel('Cycling: \n Percentage of a Equilvalent Day or Week (%)', color='purple')
ax1.tick_params(axis='y', labelcolor='purple')

ax2 = ax1.twinx()

ax2.plot(stringency_data['Date'], stringency_data['Stringency'])
ax2.set_ylabel('Stringency Index in the UK (0-100)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

fig.tight_layout()
plt.title("Cycling in the UK since 1st March 2020 vs Stringency Index in the UK")
plt.axvline(x = np.datetime64('2020-03-23'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-02-19'), -2, '23rd March 2020: \n Lockdown Starts')

plt.axvline(x = np.datetime64('2020-05-13'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-05-14'), -2, '1st Lockdown Easing')

plt.axvline(x = np.datetime64('2020-06-01'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-06-02'), 0, '2nd Lockdown Easing')

plt.axvline(x = np.datetime64('2020-06-15'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-06-16'), 2, '3rd Lockdown Easing')

plt.axvline(x = np.datetime64('2020-07-04'), color='black', linestyle = '--')
plt.text(np.datetime64('2020-07-05'), 4, '4th Lockdown Easing')

plt.show()
```
